```python
import json
import glob
dir_path = "../data/**/*addRailname.geojson"
geojson = glob.glob(dir_path, recursive=True)
for file in geojson:
    # print(file)
    new_file = file[:-19]+"station_only.geojson"
    print(new_file)
    station_only = {}
    station_only["type"] = "FeatureCollection"
    station_only["features"] = []
    with open(file, "r") as j:
        a = json.load(j)
    for element in a["features"]:
        if element["geometry"]["type"] == "Point":
            # print(element)
            station_only["features"].append(element)
    with open(new_file, "w")as k:
        json.dump(station_only, k, ensure_ascii=False)
    
# with open("./station_only.geojson","w") as j:
#     json.dump(station_only, j, ensure_ascii=False)
```
